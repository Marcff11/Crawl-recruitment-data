```python
header = 'Stt	Ngày nhận hồ sơ	Ngày nhận HS 	Tháng nhận HS	Năm nhận HS	Công ty	Đơn vị / Chi nhánh	Khối/TT	Phòng ban/Bộ phận	Vị trí	Họ và tên	Ngày tháng năm sinh	Số điện thoại	Email	Tuổi	Giới tính	Tốt nghiệp trường	Trình độ cao nhất	Địa chỉ	Nguồn	Hình thức	Loại kênh	Kênh TD	KQ hồ sơ'
header = header.split('\t')
df = pd.DataFrame(columns = header)
df
```

```python
phone = re.findall(r"[0|84|+84][\d\._\- ]{7,}",s)
email = re.findall(r"[\w_\.\-]+@[A-Za-z]+\.[A-Za-z\.]+",s)
applying_date = re.findall(r"\d\d/\d\d/\d\d\d\d",s)
name = re.findall(r"(.+)\n.+\n.+\n#",s)
position = re.findall(r"(.+)\n#",s)
status = re.findall(r"CV tiếp nhận|Phù hợp|Hẹn phỏng vấn|Gửi đề nghị|Nhận việc|Từ chối",s)
phone = ["'" + i for i in (phone)]
[print(i) for i in [phone, email, applying_date, name, position,status]]
```
